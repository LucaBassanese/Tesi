# 2d scaling

```
jacc = pd.read_excel('/content/jaccardpes.xlsx', index_col=0)
```

```
a = jacc.values
np.fill_diagonal(a, 0)
df = pd.DataFrame(a)
```

```
df.index = jacc.index
df.columns = jacc.index
```

```
massimo = df.max().max()
```

```
df = df / massimo
```

```
a = df.values
np.fill_diagonal(a, 1)
df = pd.DataFrame(a)
```

```
df.index = jacc.index
df.columns = jacc.index
```

```
dissim = 1-df
```

```
model = MDS(n_components=2, dissimilarity='precomputed', random_state=1)
smd = model.fit_transform(dissim)
```

```
risul = pd.DataFrame(smd)
risul['Autore'] =  list(jacc.index)
```

```
risul
```

```
finale = pd.merge(risul, res, on= 'Autore', how='left')
```

```
risul.to_excel('scaling2d.xlsx', index= False)
```
