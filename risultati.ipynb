```
a = esiti['Relazione tra keywords'].map({'Non correlate' : 0, 'Debolmente correlate': 0.25,'Correlate' : 0.5, 'Fortemente correlate':0.75, 'Sinonimi':1 })
```

```
esiti['Valori'] = a
```

```
matriciona_key = pd.read_excel('/content/Tesi/data/matriciona_key.xlsx', index_col= 0)
```

```
spearman_corr = []
pvalue_sp = []
pearson_corr = []
pvalue_pr = []
for i in range(len(matriciona_key)):
  spearman_corr.append(stats.spearmanr(esiti.Valori, matriciona_key.iloc[i])[0])
  pvalue_sp.append(stats.spearmanr( esiti.Valori, matriciona_key.iloc[i])[1])
  pearson_corr.append(stats.pearsonr(esiti.Valori, matriciona_key.iloc[i])[0])
  pvalue_pr.append(stats.pearsonr(esiti.Valori, matriciona_key.iloc[i])[1])
```

```
dizio = {'Correlazione Spearman': spearman_corr, 'pvalue Spearman':pvalue_sp, 'Correlazione Pearson': pearson_corr, 'pvalue Pearson': pvalue_pr}
```

```
valuta = pd.DataFrame(dizio)
```

```
valuta.index = matriciona_key.index
```
