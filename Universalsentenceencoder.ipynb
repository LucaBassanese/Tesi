```
#@title Load the Universal Sentence Encoder's TF Hub module
from absl import logging

import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" #@param ["https://tfhub.dev/google/universal-sentence-encoder/4", "https://tfhub.dev/google/universal-sentence-encoder-large/5"]
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

#Universal

```
import pandas as pd
frasi = pd.read_csv('/content/Tesi/data/frasi.txt', header=None)
```

```
frasi.columns = ['titolo']
```

```
# Reduce logging output.
logging.set_verbosity(logging.ERROR)
message_embeddings = embed(list(frasi.titolo))
```

```
matrix = np.zeros((len(frasi),512))
for i in range(len(frasi)):
  matrix[i,:] = message_embeddings[i]
```

```
a = pd.DataFrame(matrix)
a.to_csv('/content/Tesi/data/universal.csv', index= False)
```
